Predicting whether LendingClub borrowers (2007-2010) fully repaid their loans.

## Setting Up the Data
**Create dummy variables for the `purpose` column**

```python
cat_feats = ['purpose']
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
final_data.info()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
print(f'Training: {X_train.shape[0]} rows | Test: {X_test.shape[0]} rows')
```

## Training a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

### Predictions and Evaluation – Decision Tree

```python
from sklearn.metrics import classification_report, confusion_matrix

dt_pred = dtree.predict(X_test)
print('Confusion Matrix:')
print(confusion_matrix(y_test, dt_pred))
print()
print(classification_report(y_test, dt_pred))
```

## Training the Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200, random_state=101)
rfc.fit(X_train, y_train)
```

### Predictions and Evaluation – Random Forest

```python
rfc_pred = rfc.predict(X_test)
print('Confusion Matrix:')
print(confusion_matrix(y_test, rfc_pred))
print()
print(classification_report(y_test, rfc_pred))
```

### Which performed better?

| Model | Accuracy | Class-1 Recall | Class-1 Precision |
|---|---|---|---|
| Decision Tree | 73% | 24% | 19% |
| Random Forest | 85% | 2% | 45% |

- **Overall accuracy**: Random Forest wins (85% vs 73%).
- **Detecting non-repayment**: Decision Tree actually finds more bad loans (recall 24% vs 2%), though with low precision.
- The dataset is **class-imbalanced** (~84% fully paid), so the Random Forest learned to predict "fully paid" almost always.
- To improve: try `class_weight='balanced'`, SMOTE, or threshold tuning.

### Feature Importances

```python
importances = pd.Series(rfc.feature_importances_, index=X.columns).sort_values(ascending=False)
importances[:12].plot(kind='bar', figsize=(10,4), color='steelblue', alpha=0.85)
plt.title('Top 12 Feature Importances – Random Forest')
plt.ylabel('Importance'); plt.tight_layout(); plt.show()
```
